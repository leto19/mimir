```python
import wikipedia, nltk, re
```

```python
nltk.download('punkt')
nltk.download('averaged_perceptron_tagger')
nltk.download('maxent_ne_chunker')
nltk.download('words')
```

```python
page = wikipedia.page("A Game of Thrones", auto_suggest=False)
```

```python
content = page.content
```

```python
# regex = re.compile('[^a-zA-Z]')
# words = regex.match(content)
# print(words)
```

```python
tokens = nltk.word_tokenize(content)
tokens = nltk.pos_tag(tokens)
```

```python
entities = nltk.chunk.ne_chunk(tokens)
print(entities[0:20])
```

```python
pattern = 'NP: {<DT>?<JJ>*<NN>}'
cp = nltk.RegexpParser(pattern)
cs = cp.parse(entities)
print(cs)
```

```python
from nltk.chunk import conlltags2tree, tree2conlltags
from pprint import pprint

iob_tagged = tree2conlltags(cs)
pprint(iob_tagged)
```

```python
peeps = [x for x in iob_tagged if (x[2] == "B-PERSON" or x[2] == "I-PERSON")]
print(peeps)
```
