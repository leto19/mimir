```
import pandas as pd
```

```
from google.colab import drive
drive.mount('/content/gdrive')

ROOT        = 'gdrive/Shared drives/CDT Mini-Project Team 1/Colab Notebooks/'
DATA_DIR    = ROOT + 'data/' 
MODELS_DIR  = ROOT + 'models/'
```

```
#Load all data

qaps = pd.read_csv(DATA_DIR + 'narrativeqa_qas.csv')
summaries = pd.read_csv(DATA_DIR + 'summaries.csv')
```

```
summaries.head()
```

```
qaps.head()
```

```
start_idxs = []
answers = []
list_summaries = []

for index, row in qaps.iterrows():
  doc_id = row['document_id']
  summary = summaries[summaries['document_id']==doc_id]['summary_tokenized'].iloc[0]
  question = row['question_tokenized']
  answer1 = row['answer1_tokenized']
  answer2 = row['answer2_tokenized']

  #some answers don't appear because they were written with a full-stop at the end - remove these
  if answer1.endswith('.'):
    answer1 = answer1[:-1]
  if answer2.endswith('.'):
    answer2 = answer2[:-1]

  answer = answer1
  start_idx = summary.find(answer1)
  if start_idx == -1:
    start_idx = summary.find(answer2)
    answer = answer2

  start_idxs.append(start_idx)
  answers.append(answer)
  list_summaries.append(summary)
```

```
qaps = qaps.drop(columns=['document_id', 'question','answer1','answer2', 'answer1_tokenized','answer2_tokenized'],axis=1)
qaps['summary'] = list_summaries
qaps['answer'] = answers
qaps['ans_start'] = start_idxs
```

```
qaps.sample(20)
```

```
qaps.to_csv(DATA_DIR + 'narrativeqa_qaps_single_answer.csv')
```
